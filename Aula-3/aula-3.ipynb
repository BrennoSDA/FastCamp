```python
#Importando as biblitecas necessárias
from enum import auto, IntFlag
from typing import Any

from pydantic import(
    BaseModel,
    EmailStr,
    Field,
    SecretBytes,
    ValidationError,
)

#Definicao das roles e definindo que o admin tem privilegios de todas
class role(IntFlag):
    Author = auto()
    Editor = auto()
    Developer = auto()
    Admin = Author | Editor | Developer

#Definindo o usuario e todos os campos que o mesmo terá, com definicao, exemplo e se é multavel ou nao 
class user(BaseModel):
    name: str = Field(examples = ["Arjan"])
    email: EmailStr = Field(
        examples = ["example@arjancodes.com"], #Exemplo de e-mail válido
        description = "the email address of the user", #Definicao do que é o e-mail
        frozen = True, #Significa que o e-mail nao e multavel
    )
    password: SecretStr = Field(
        examples = ["Password123"], description = "The password of the user"
    )
    role: Role = Field(default = None, description = "The role of the user")

#Funcao de validacao que irá validar os campos de uma vez e caso tenha algum erro retornara o mesmo 
def validate (data: dict[str, Any]) -> None:
    try:
        user = User.model_validate(data)
        print(user)
    except ValidationError as e:
        print("User is invalid")
        for error in e.erros():
            print(error)

#Funcao main para testar, foi fornecido uma good_data, ou seja, informacoes de acordo com o esperado e testado, nao houve nenhum erro 
def main() -> None:
    good_data = {
        "name": "Arjan", 
        "email": "example@arjancodes.com",
        "password": "Password123"
        }

    bad_data = {"email": "<bad data>", "password": "<bad data>"}

    validate(good_data)
    validate(bad_data)
```
